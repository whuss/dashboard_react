```python
import numpy as np
import pandas as pd
from datetime import datetime, timedelta
from app import db
from db import DeviceInfo

from db import TemperaturePackage, HumidityPackage, PressurePackage, GasPackage, BrightnessPackage
```

```python
device = 'PTL_RD_AT_001'
```

```python
since = datetime.strptime('2020-03-22 00:00:00', '%Y-%m-%d %H:%M:%S')
until = datetime.strptime('2020-03-22 23:59:59', '%Y-%m-%d %H:%M:%S')
```

```python
tp = TemperaturePackage
query = db.session.query(tp.temperature) \
    .filter(tp.device == device) \
    .filter(tp.timestamp >= since) \
    .filter(tp.timestamp <= until)
tdata = pd.DataFrame(query.limit(1000).all())
tdata.size
```

```python
hp = HumidityPackage
query = db.session.query(hp.humidity) \
    .filter(hp.device == device) \
    .filter(hp.timestamp >= since) \
    .filter(hp.timestamp <= until)
hdata = pd.DataFrame(query.limit(1000).all())
hdata.size
```

```python
pp = PressurePackage
query = db.session.query(pp.pressure) \
    .filter(pp.device == device) \
    .filter(pp.timestamp >= since) \
    .filter(pp.timestamp <= until)
pdata = pd.DataFrame(query.limit(1000).all())
pdata.size
```

```python
gp = GasPackage
query = db.session.query(gp.amount) \
    .filter(gp.device == device) \
    .filter(gp.timestamp >= since) \
    .filter(gp.timestamp <= until)
gdata = pd.DataFrame(query.limit(1000).all()) \
    .rename(columns=dict(amount="gas"))
gdata.size
```

```python
bp = BrightnessPackage
query = db.session.query(bp.brightness) \
    .filter(bp.device == device) \
    .filter(bp.source == 'brightness_l_h@BH1750') \
    .filter(bp.timestamp >= since) \
    .filter(bp.timestamp <= until)
b_lh_data = pd.DataFrame(query.limit(1000).all()) \
    .rename(columns=dict(brightness="brightness_l_h"))
b_lh_data.size
```

```python
bp = BrightnessPackage
query = db.session.query(bp.brightness) \
    .filter(bp.device == device) \
    .filter(bp.source == 'brightness_l_v@BH1750') \
    .filter(bp.timestamp >= since) \
    .filter(bp.timestamp <= until)
b_lv_data = pd.DataFrame(query.limit(1000).all()) \
    .rename(columns=dict(brightness="brightness_l_v"))
b_lv_data.size
```

```python
bp = BrightnessPackage
query = db.session.query(bp.brightness) \
    .filter(bp.device == device) \
    .filter(bp.source == 'brightness_r_h@BH1750') \
    .filter(bp.timestamp >= since) \
    .filter(bp.timestamp <= until)
b_rh_data = pd.DataFrame(query.limit(1000).all()) \
    .rename(columns=dict(brightness="brightness_r_h"))
b_rh_data.size
```

```python
bp = BrightnessPackage
query = db.session.query(bp.brightness) \
    .filter(bp.device == device) \
    .filter(bp.source == 'brightness_r_v@BH1750') \
    .filter(bp.timestamp >= since) \
    .filter(bp.timestamp <= until)
b_rv_data = pd.DataFrame(query.limit(1000).all()) \
    .rename(columns=dict(brightness="brightness_r_v"))
b_rv_data.size
```

```python
sensor_data = pd.concat([tdata, hdata, pdata, b_lh_data, b_lv_data, b_rh_data, b_rv_data, gdata], axis = 1)
sensor_data
```

```python
sensor_data.to_csv("sensor_data.csv", index=False)
```

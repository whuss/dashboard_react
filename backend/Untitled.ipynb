```python
import numpy as np
import pandas as pd

number_of_points = 100
# create three normal population samples with different parameters
x1 = np.random.normal(loc=5.0, size=number_of_points*2) * 100
y1 = np.random.normal(loc=10.0, size=number_of_points*2) * 10

x2 = np.random.normal(loc=5.0, size=number_of_points*4) * 50
y2 = np.random.normal(loc=5.0, size=number_of_points*4) * 10

x3 = np.random.normal(loc=55.0, size=number_of_points) * 10
y3 = np.random.normal(loc=4.0, size=number_of_points) * 10

x = np.concatenate((x1, x2, x3))
y = np.concatenate((y1, y2, y3))
```

```python
np.vstack([x,y]).shape
```

```python
data = pd.DataFrame(np.vstack([x,y]).T, columns=['x', 'y'])
```
